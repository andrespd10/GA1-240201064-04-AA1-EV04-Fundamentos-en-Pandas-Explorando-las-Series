## Cargar desde un CSV

```python
import pandas as pd
import csv

# Usamos parámetros extra para manejar el error de la fila 18303
try:
    df = pd.read_csv(
        '../data/raw/incidentes_viales.csv', 
        encoding='utf-8',        # Para tildes y eñes
        sep=None,                  # Detecta si es coma o punto y coma
        engine='python',           # Motor más flexible para errores
        quoting=csv.QUOTE_NONE,    # Ignora comillas problemáticas
        on_bad_lines='warn'        # Salta líneas corruptas y avisa
    )
    print("¡Carga exitosa! Dataset listo para el taller.")
    display(df.head())             # Punto 4 de tu evidencia
except Exception as e:
    print(f"No se pudo cargar: {e}")
```
