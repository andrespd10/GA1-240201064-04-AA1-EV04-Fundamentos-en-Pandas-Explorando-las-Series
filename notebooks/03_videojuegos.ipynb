## Resumen Estadístico:

```python
df.describe()##: Para columnas numéricas, te da la media, mediana (50%), desviación estándar, mínimo, máximo y cuartiles. Es excelente para detectar outliers (valores atípicos) a simple vista (ej. una 'Edad' de 200 años o un 'Precio' negativo).
```

```python
df.describe(include='object') ##: Para columnas de texto (object), te da el conteo, el número de valores únicos (unique), el valor más común (top) y su frecuencia (freq).
```
